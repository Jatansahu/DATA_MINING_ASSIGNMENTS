```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://edition.cnn.com/2023/08/03/business/india-rice-export-ban/index.html"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
# Get the title
title = soup.title
print(title)
```

```python
updtime = soup.find(class_='timestamp')
print(updtime.getText())
```

```python
article_txt = soup.find(class_='article__content')
print(article_txt.getText())
```
